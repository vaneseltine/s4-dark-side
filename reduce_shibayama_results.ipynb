```python
import json
from pathlib import Path
```

```python
in_paths = list(Path('D:/').glob('result_*.json'))
OUT_DIR = Path('./data/shibayama/')
out_paths = list(OUT_DIR / f'bare_{path.name}' for path in in_paths)
list(zip(in_paths, out_paths))
```

```python
def make_bare(records):
    bare = {
        x['id'].replace('https://openalex.org/', ''): (
            x['shibayama']['shibayama_title_embedding']['100%'], 
            x['shibayama'].get('shibayama_abstract_embedding', {}).get('100%', None)
        ) 
        for x
        in records
    }
    return bare
```

```python
all_records =[]
for in_path, out_path in zip(in_paths, out_paths):
    print(in_path, end=', ')
    with in_path.open('rb') as infile:
        new_records = json.load(infile)
    all_records += new_records
    bare = make_bare(new_records)
    print(len(bare))
    with out_path.open('w') as outfile:
        json.dump(bare, outfile)
    print(out_path, round(out_path.stat().st_size / 1024 / 1024, 1), 'mb')
len(all_records)
```
